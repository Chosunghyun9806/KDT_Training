```python
from PIL import Image
import glob
import os

dir_ = './vehicle/train/images'
save_path = './vehicle/train/resize/'
load_paths = glob.glob(dir_ + '/*.jpg')
token_list = os.listdir(dir_)

for i in range(len(load_paths)):
    im = Image.open(load_paths[i])
    save = (save_path + token_list[i])
    print(save)
    im = im.resize((512, 512))
    im = im.convert('RGB')
    im.save(save)
    
print('resize complete')
```
