```python
import glob
import os
import string

dir_ = './vehicle/valid/labels'
load_paths = glob.glob(dir_ + '/*.txt')
token_list = os.listdir(dir_)
```

```python
for i in range(len(load_paths)):
    fr = open(load_paths[i], 'r')
    lines = fr.readlines()
    a = (''.join(lines))
    fr.close()
    
    fw = open(load_paths[i], 'w')
    for line in lines:
        if a.startswith('4'):
            fw.write(line.replace('4','10',1))
        elif a.startswith('2'):
            fw.write(line.replace('2','1',1))
        elif a.startswith('0'):
            fw.write(line.replace('0','11',1))
        elif a.startswith('1'):
            fw.write(line.replace('1','12',1))
        elif a.startswith('3'):
            fw.write(line.replace('3','13',1))     
        else:
            continue
            
    fw.close()
```
