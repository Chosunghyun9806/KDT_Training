```python
import requests
from bs4 import BeautifulSoup

url = "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&pkid=68&os=10105426&qvt=0&query=%EB%B9%84%EC%83%81%EC%84%A0%EC%96%B8"
res = requests.get(url)
res.raise_for_status()
```

```python
dt = BeautifulSoup(res.text, 'html.parser')
```

```python
images = dt.find("div" , attrs = {"class" : "type_plural list_exact movie_list"})
images2 = images.find_all("img")
```

```python
print(images2[0])
```

```python
print(images2[0]['alt'])
```

```python
for i,image in enumerate(images2):
    img_url = images2['src']
    if img_url.startswith("//"):
        img_url = "https:" + img_url
    
    img_res = requests.get(img_url)
    img_res.raise_for_status()

    with open('movie().jpg'.format(i+1), 'wb') as f:
        f.write(img_res.content)
```

```python
for idx, image in enumerate(images2):
    print(images2)
    img_url = image['src']
    if img_url.startswith("//"):
        img_url = "https:" + img_url
    img_res = requests.get(img_url)
    # 오류 검토
    img_res.raise_for_status()
    with open("./image/movie{}.jpg".format(idx+1), "wb") as f:
        f.write(img_res.content)
```
