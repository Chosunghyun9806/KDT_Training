```python
import requests
import bs4

res = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%82%A0%EC%94%A8")

if res.status_code == requests.codes.ok:
    print('정상')
else:
    print('오류 : %d' %res.status_code)
    
bs_obj = bs4.BeautifulSoup(res.content, 'lxml')
atag = bs_obj.find('div', {'class':'temperature_info'})
print(atag)
```

```python
import requests
import bs4

res = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%82%A0%EC%94%A8")

if res.status_code == requests.codes.ok:
    print('정상')
else:
    print('오류 : %d' %res.status_code)
    
bs_obj = bs4.BeautifulSoup(res.content, 'lxml')
atag = bs_obj.findAll('dd', {'class':'desc'})
print(atag)
```
