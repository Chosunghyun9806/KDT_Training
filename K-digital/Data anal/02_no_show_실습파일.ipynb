### 2-1-3 One Hot encoding 1(get_dummies)

`Neighbourhood` 칼럼을 One Hot encoding 해보겠습니다.


<b> one-hot encoding</b> 이란

범주형 변수를 숫자로 바꿀 때 개별로 목록을 만들고 이진(0/1)값을 넣는 방법

예를 들어 위 칼럼은 ['AEROPORTO', 'ANDORINHAS', 'ANTÔNIO HONÓRIO',
       'ARIOVALDO FAVALESSA', 등등] 있었으므로 
       아래처럼 표기할 수 있습니다.

AEROPORTO | ANDORINHAS | ANTÔNIO HONÓRIO |ARIOVALDO FAVALESSA
---|---|---|---
0|1|0|0
1|0|0|0

칼럼에 unique한 값이 많아질수록 데이터의 차원 수가 증가하게 됩니다.

Encoding시 몇가지 주의해야할 점이 있습니다.

일부 데이터의 경우 train 에는 존재하는 label이 test에는 없거나 그 반대인 경우가 있습니다.

이런 경우 One hot encoding을 사용하면 데이터의 차원이 맞지 않아서 test 데이터의 분석시 오류가 발생합니다.

train에 존재하지만 test에 없는 라벨의 경우 test에 0으로 채운 컬럼을 추가하여 사용합니다.

반대의 경우에도 train에 0으로 채운 컬럼을 추가하여 train과 test set의 shape을 맞춰줍니다.
<br><br>


<b>
One-Hot encoding을 도와주는 함수는 두 가지가 있습니다.

- pandas 의 get_dummies
- sklearn.preprocessing 의 OneHotEncoder

</b>
