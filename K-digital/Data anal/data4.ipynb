```python
df_can = df[df['사망원인별(103항목)'] == '신생물 (C00-D48)']
df_can
df_canm = df[
    (df['사망원인별(103항목)'] == '신생물 (C00-D48)') &
           (df['성별'] == '남자')]
df_canf = df[
    (df['사망원인별(103항목)'] == '신생물 (C00-D48)') &
           (df['성별'] == '여자')]

print(df_canm.head())
print('\n')
print(df_canf.head())
df_canmt = df_canm.set_index('시도별')
df_canft = df_canf.set_index('시도별')

df_canmt[1:].plot(kind='bar', figsize=(10,6), title = '시도별 사망자수(남자)')
```
