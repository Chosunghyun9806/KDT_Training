```python
import pandas as pd
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

df = pd.read_csv('./지역별의원.csv', header = 6, encoding='cp949')

df['영상의학과'].fillna(0,inplace=True)
hos_df = df[['내과','마취통증의학과','영상의학과','외과','시도']]
hos = hos_df.set_index(['시도'])

hos[1:].plot(kind='bar', figsize=(10,5), title = '지역별 의원 수')
```
