```python
import pandas as pd

student1 = pd.Series({'국어':100,'영어':80 ,'수학':90})
student2 = pd.Series({'수학':80, '국어':90, '영어':80})
print(student1)
print('\n')
print(student2)
print('\n')

addtion = student1 + student2
subtraction = student1 - student2
multipication = student1 * student2
division = student1/student2
print(type(division))
print('\n')

result = pd.DataFrame([addtion, subtraction, multipication, division],
                      index = ['덧셈','뺄셈','곱셈','나눗셈'])
print(result)
```

```python
import pandas as pd
import numpy as np

student1 = pd.Series({'국어':np.nan,'영어':80 ,'수학':90})
student2 = pd.Series({'수학':80, '국어':90, '영어':80})
print(student1)
print('\n')
print(student2)
print('\n')

sr_add = student1.add(student2, fill_value=0)
sr_sub = student1.sub(student2, fill_value=0)
sr_mul = student1.mul(student2, fill_value=0)
sr_div = student1.div(student2, fill_value=0)

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                     index = ['덧셈','뺄셈','곱셈','나눗셈'])
print(result)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
sns.set_style('darkgrid')
fig = plt.figure(figsize=(15,5)) # 가로 15 세로 5 사각형 세팅
ax1 = fig.add_subplot(1,2,1) # 나눠서 1번에 ax1 대입
ax2 = fig.add_subplot(1,2,2) # 나눠서 2번에 ax2 대입

sns.regplot(x='age',
           y='fare',
           data=titanic,
           ax=ax1)

sns.regplot(x='age',
           y='fare',
           data=titanic,
           ax=ax2,
           fit_reg=False)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
fig = plt.figure(figsize=(15,5)) # 가로 15 세로 5 사각형 세팅
ax1 = fig.add_subplot(1,3,1) # 나눠서 1번에 ax1 대입
ax2 = fig.add_subplot(1,3,2) # 나눠서 2번에 ax2 대입
ax3 = fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'],ax=ax1)
sns.distplot(titanic['fare'],hist=False,ax=ax2)
sns.distplot(titanic['fare'],kde=False, ax=ax3)

ax1.set_title('titanic fare - hist/ked')
ax2.set_title('titanic fare - ked')
ax3.set_title('titanic fare - hist')

plt.show()
```
