## 분석과 시각화를 위한 tidy data 만들기

#### https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf

```python
df_1 = df.melt(id_vars =['국가(대륙)별','상품군별','판매유형별'],var_name ='기간', value_name ='million')
df_1
          
```

```python
df_1['연도'] = df_1['기간'].map(lambda x : x.split('.'[0]))
```

```python
df_1
```

```python
df_1['분기'] = df_1['기간'].apply(lambda x : x.split('.')[1].split('/')[0])
```

```python
df_1
```

```python
df_1['분기'] = df_1['분기'].astype(int)
df_1.head()
```

### 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
df_1['million'] = df_1['million'].replace('-', pd.np.nan).astype(float)
df_1
```

#### 필요없는 데이터 제거하기
