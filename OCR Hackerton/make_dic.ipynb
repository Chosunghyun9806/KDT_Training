```python
from PIL import Image
import glob
import os
import shutil
import itertools
import numpy as np
import json
import pandas as pd
```

```python
dir_1 = r'./OCR_Hack/sum_data/json_standard'
json_dir = glob.glob(dir_1 + './*.json')

json_dir[1]
```

```python
classes = []

for i in range(len(json_dir)):
    with open(json_dir[i], 'r', encoding = 'utf-8') as file:
        data = json.load(file)
        A = data['Image_Text_Coord']
        for j in range(len(A)):
            for k in range(len(A[j])):
                B = A[j][k]['label']
                classes.append(B)
                
                
classes = list(set(classes))
print(classes)
print('\n')
print('클래스 타입 : %s'  %(type(classes)))
print('\n')
print('class 갯수 : %s' %(len(classes)))
```

```python
dic = {}

for i in range(len(classes)):
    dic[classes[i]] = i+1
```
