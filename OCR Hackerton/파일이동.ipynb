```python
dir_ = '.\OCR_Hack\sum_data\sum_json'
save_dir = '.\OCR_Hack\sum_data\json_standard\\'
output = glob.glob(dir_ + '/*.json')
```

```python
dir_img = '.\OCR_Hack\sum_data\images'
save_dir_img = '.\OCR_Hack\sum_data\images_reshape\\'
output_img = glob.glob(dir_img + '/*.jpg')
```

```python
output[0]
print(output[0].split('\\')[4])
```

```python
print(output_img[0])
```

```python
for i in range(len(output_img)):
    #print(output_img[i])
    img = cv2.imread(output_img[i])
    YY = np.round((img.shape[0] / 320), 5)
    XX = np.round((img.shape[1] / 256), 5)
    print(i)
    #print(XX,YY)
    with open(output[i], 'r', encoding = 'utf-8') as file:
        #print(output[i])
        data = json.load(file)
        A = data['Image_Text_Coord']
        for j in range(len(A)):
            for k in range(len(A[j])):
                B = A[j][k]
                C = B['bbox']
                x1 = C[0] / XX
                y1 = C[1] / YY
                x2 = C[2] / XX
                y2 = C[3] / YY
                C_0 = x1 / 256
                C_1 = y1 / 320
                C_2 = x2 / 256
                C_3 = y2 / 320
                #print(C_0, C_1, C_2, C_3)
                B.update({'bbox' : [C_0, C_1, C_2, C_3, C[4], C[5]]})
    with open(save_dir + (output[i].split('\\')[4]), 'w', encoding = 'utf-8') as outfile:
        json.dump(data, outfile)
    
    img = cv2.resize(img, dsize = (256, 320))
    cv2.imwrite(save_dir_img + (output_img[i].split('\\')[4]) , img)
        
print('Complete')
```

```python
dir_json2 = '.\OCR_Hack\sum_data\json_standard'
json_path = glob.glob(dir_json2 + './*.json')

dir_img2 = '.\OCR_Hack\sum_data\images'
image_path = glob.glob(dir_img2 + './*.jpg')
```

```python
print(json_path[0])
print(image_path[0])
```

```python
for o in range(3):
    with open(json_path[o], 'r', encoding = 'utf-8') as f:
        json_data = json.load(f)
        
        coord_ls = json_data['Image_Text_Coord']

        bbox_ls = []
        for i in range(len(coord_ls)):
            for j in range(len(coord_ls[i])):
                bbox = coord_ls[i][j]['bbox']
                bbox_ls.append(bbox)
        
        img = cv2.imread(image_path[o], 1) # color로 읽어오기
        resize_img = cv2.resize(img, dsize = (256, 320))
        
        bbox_img = resize_img.copy()
        for l in range(len(bbox_ls)):
            x = bbox_ls[l][0]
            y = bbox_ls[l][1]
            w = bbox_ls[l][2]
            h = bbox_ls[l][3]
            
            bbox_img = cv2.rectangle(bbox_img, (int(x * 256), int(y * 320)), (int((x + w) * 256), int((y + h) * 320)), (255,0,0))
            
        cvt_bbox_img = cv2.cvtColor(bbox_img, cv2.COLOR_BGR2RGB)
        plt.imshow(cvt_bbox_img)
        plt.show()
        
        #return bbox_img
```
