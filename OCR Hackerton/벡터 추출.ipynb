```python
from PIL import Image
import glob
import os
import shutil
import itertools
import numpy as np
import json
```

```python
dir_1 = r'./OCR_Hack/sum_data/json_standard'
json_dir = glob.glob(dir_1 + './*.json')

print(json_dir)
```

```python
classes = []

for i in range(len(json_dir)):
    with open(json_dir[i], 'r', encoding = 'utf-8') as file:
        data = json.load(file)
        A = data['Image_Text_Coord']
        for j in range(len(A)):
            for k in range(len(A[j])):
                B = A[j][k]['label']
                classes.append(B)
                
                
classes = list(set(classes))
print(classes)
print('\n')
print('  : %s'  %(type(classes)))
print('\n')
print('class  : %s' %(len(classes)))
```
